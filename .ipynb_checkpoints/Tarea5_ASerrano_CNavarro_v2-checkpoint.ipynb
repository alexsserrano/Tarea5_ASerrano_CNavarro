# Pregunta 6
> 6. Comparar el porcentaje de puntos de aceptación cuando se usa $t(x)$ constante y $t(x)$ un pulso senoidal. Concluir

```python
#Pregunta 6 Codigo Carola
def aceptacion_rechazo(N, f, t, x_min, x_max):
    # Encuentra el máximo de f en el rango dado
    max_f = f(optimize.fmin(lambda x: -f(x), (x_max - x_min) / 2, disp=False))

    # Genera puntos aleatorios en el rango dado
    X = np.random.uniform(x_min, x_max, N)

    # Genera puntos aleatorios en el rectángulo t(x) x max_f
    U = np.random.uniform(0, max_f, N)

    # Evalúa f(x) y t(x) en los puntos aleatorios
    fx = f(X)
    tx = t(X)

    # Evalúa la condición de aceptación y rechazo
    accept = U < fx / tx

    # Grafica los puntos aceptados y rechazados
    plt.plot(X[accept], U[accept], 'o', alpha=0.5, label='Aceptados')
    plt.plot(X[~accept], U[~accept], 'o', alpha=0.5, label='Rechazados')
    plt.plot(np.arange(x_min, x_max, 0.01), t(np.arange(x_min, x_max, 0.01)), 'k-', label='t(x)')
    plt.xlabel('x')
    plt.ylabel('f(x) / t(x)')
    plt.legend()
    plt.show()

# Ejemplo 1: t(x) constante
f = lambda x: 30 * (x**2 - 2*x**3 + x**4)
t = lambda x: np.ones_like(x) * max_f
x_min, x_max = 0, 1
N = 10000
aceptacion_rechazo(N, f, t, x_min, x_max)

# Ejemplo 2: t(x) un pulso senoidal
f = lambda x: 30 * (x**2 - 2*x**3 + x**4)
t = lambda x: max_f * np.sin(np.pi * x)
x_min, x_max = 0, 1
N = 10000
aceptacion_rechazo(N, f, t, x_min, x_max)
```

```python
# Codigo Alex

# Función objetivo
def f(x):
    return 30 * (x**2 - 2*x**3 + x**4)

# Máximo de la función f
max_f = f(optimize.fmin(lambda x:-f(x), 0, disp=False))

# Función t constante
def t_constante(x):
    return max_f * np.ones_like(x)

# Función t senoidal
def t_senoidal(x):
    return max_f * np.sin(np.pi * x)

# Función de aceptación y rechazo que devuelve el porcentaje de puntos aceptados
def porcentaje_aceptacion(R2, R1, f, t):
    f_x = f(R1)
    t_x = t(R1)
    condition = R2 * t_x <= f_x
    return 100 * np.sum(condition) / len(R1)

N = 10000  # número de puntos a simular
R2 = np.random.rand(N)
R1 = np.random.rand(N)

# Porcentajes de aceptación
porcentaje_constante = porcentaje_aceptacion(R2, R1, f, t_constante)
porcentaje_senoidal = porcentaje_aceptacion(R2, R1, f, t_senoidal)

print(f"Porcentaje de aceptación con t(x) constante: {porcentaje_constante:.2f}%")
print(f"Porcentaje de aceptación con t(x) senoidal: {porcentaje_senoidal:.2f}%")
```

# Pregunta 8
> 8. Demostrar la validez de la siguiente ecuación
$$p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i$$



## Carola

$$p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i$$

Cuando $i=0$, la ecuación se reduce a:

$$p_{i+1}=\frac{n-0}{0+1}\frac{p}{1-p} p_0 = \frac{n}{1}\frac{p}{1-p} p_0$$ 


## Alex:


Para demostrar la validez de la ecuación dada, vamos a definir algunos términos:

- $p$: Probabilidad de éxito.
- $1-p$: Probabilidad de fracaso.
- $p_i$: Probabilidad de tener exactamente $i$ éxitos en $n$ ensayos.

La ecuación dada parece relacionar la probabilidad de tener $i+1$ éxitos, $p_{i+1}$, con la probabilidad de tener $i$ éxitos, $p_i$.

**Demostración**:

La probabilidad de tener $i+1$ éxitos en $n$ ensayos se puede ver como la probabilidad de tener $i$ éxitos en los primeros $n-1$ ensayos y un éxito en el ensayo $n$, más la probabilidad de tener $i+1$ éxitos en los primeros $n-1$ ensayos y un fracaso en el ensayo $n$.

1. La probabilidad de tener $i$ éxitos en los primeros $n-1$ ensayos y un éxito en el ensayo $n$ es:
$$ p_i \times p $$

2. La probabilidad de tener $i+1$ éxitos en los primeros $n-1$ ensayos y un fracaso en el ensayo $n$ es:
$$ \binom{n-1}{i+1} p^{i+1} (1-p) $$
Pero, $\binom{n-1}{i+1}$ es el número de formas de elegir $i+1$ éxitos de $n-1$ ensayos, que es igual a:
$$ \frac{(n-1)!}{(i+1)!(n-1-i-1)!} = \frac{n-1}{i+1} \times \frac{(n-2)!}{i!(n-2-i)!} $$
El término $\frac{(n-2)!}{i!(n-2-i)!}$ es simplemente $\binom{n-2}{i}$, que es proporcional a $p_i$. Por lo tanto, la probabilidad se convierte en:
$$ \frac{n-1}{i+1} p_i \times (1-p) $$

Sumando las dos probabilidades, obtenemos:
$$ p_{i+1} = p_i \times p + \frac{n-1}{i+1} p_i \times (1-p) $$
Factorizando $p_i$, obtenemos:
$$ p_{i+1} = p_i \left( p + \frac{n-1}{i+1} (1-p) \right) $$

Reorganizando términos, llegamos a:
$$ p_{i+1} = \frac{n-i}{i+1} \frac{p}{1-p} p_i $$

Y eso demuestra la validez de la ecuacion dada


# Pregunta 9
> 9. Seguir un procedimiento similar al mostrado cuando se generó una distribución binomial, pero en esta caso genere un código que genere variables aletorias Poisson cuya función de distribución de probabilidad esta dada por:
$$P(k,\lambda)=\frac{e^{-\lambda}(\lambda)^k}{k!}$$
    

Demuestre matemáticamente que 
 $$P(k+1)=\frac{\lambda}{k+1}P(k)$$
 y a partir de esta relación genere variables aletorias que distribuyen poisson usando el método de la transformada inversa.

## Respuesta Carola



## Respuesta Alex

### Demostración

Dada la función de probabilidad de Poisson:
$$ P(k, \lambda) = \frac{e^{-\lambda} \lambda^k}{k!} $$

Queremos demostrar que:
$$ P(k+1) = \frac{\lambda}{k+1} P(k) $$

Para demostrar esto, expresamos $P(k+1)$ en términos de $P(k)$:
$$ P(k+1) = \frac{e^{-\lambda} \lambda^{k+1}}{(k+1)!} $$

Dividimos esta expresión por $P(k)$:
$$ \frac{P(k+1)}{P(k)} = \frac{\frac{e^{-\lambda} \lambda^{k+1}}{(k+1)!}}{\frac{e^{-\lambda} \lambda^k}{k!}} $$

Simplificando:
$$ \frac{P(k+1)}{P(k)} = \frac{\lambda^{k+1} k!}{\lambda^k (k+1)!} $$
$$ \frac{P(k+1)}{P(k)} = \frac{\lambda}{k+1} $$

Multiplicando ambos lados por $P(k)$ obtenemos:
$$ P(k+1) = \frac{\lambda}{k+1} P(k) $$

Lo cual es lo que queríamos demostrar.

### Generación de Variables Aleatorias Poisson

Para generar variables aleatorias Poisson usando el método de la transformada inversa, podemos usar la relación demostrada anteriormente. El algoritmo es el siguiente:

1. Inicializar $k = 0$ y $p = e^{-\lambda}$.
2. Generar $U$ uniformemente distribuido en [0,1].
3. Mientras $U > p$:
    - $k = k + 1$
    - $p = p + \frac{\lambda}{k+1} P(k)$
4. $k$ es la variable aleatoria Poisson generada.

Este algoritmo se basa en sumar probabilidades hasta que superen a $U$. La relación demostrada nos permite calcular cada probabilidad $P(k)$ de manera eficiente sin tener que calcular factoriales o potencias en cada paso.
